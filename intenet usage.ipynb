```python
df['date']=pd.to_datetime(df['date'])
df['age_group']=df['age_group'].astype('category')
df.info()
df.head()
```

```python
df.describe()
```

```python
sns.histplot(data=df,x='age')
plt.title("Age distribution of users")
plt.show()
```

```python
purpose_cols = [
    'social_media_hours',
    'work_or_study_hours',
    'entertainment_hours'
]
sns.set_style("darkgrid")

df[purpose_cols].mean().plot(
    kind='bar',
    figsize=(8,5)
)
plt.title("Average Screen Time by Purpose")
plt.ylabel("Hours")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[purpose_cols+['total_screen_time']].corr(),annot=True,cmap='coolwarm')
plt.title("Correlation between Screen Time and Purpose")
plt.show()
```

```python
plt.figure(figsize=(7,4))
df['primary_device'].value_counts().plot(kind='bar')
plt.title("Primary Device Used")
plt.xlabel("Device")
plt.ylabel("User Count")
plt.show()
```

```python
plt.figure(figsize=(7,4))
df['internet_type'].value_counts().plot(kind='bar')
plt.title("Internet type Used")
plt.xlabel("Internet type")
plt.ylabel("User Count")
plt.show()
```

```python
sns.boxplot(data=df,x='age_group',y='total_screen_time')
plt.title("Total Screen Time by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Screen Time")
plt.show()
```

```python
daily_usage = df.groupby('date')['total_screen_time'].mean()

plt.figure(figsize=(10,5))
daily_usage.plot()
plt.title("Average Daily Screen Time Trend")
plt.xlabel("Date")
plt.ylabel("Average Screen Time")
plt.show()
```
